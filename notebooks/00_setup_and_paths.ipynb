# 00 - Setup et chemins

Objectif : vérifier l'environnement, les dépendances, les chemins du projet, les dossiers essentiels et éviter les erreurs SQLite liées à OneDrive.

Fichiers concernés : `README.md`, `requirements.txt`, `requirements-test.txt`, `src/`, `api/`, `app/`, `eval/`, `data/`, `prompts/`, `tests/`, `sql/schema.sql`.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
while not (PROJECT_ROOT / "src").exists() and PROJECT_ROOT.parent != PROJECT_ROOT:
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_DIR = PROJECT_ROOT / "data"
SAMPLE_IMAGES_DIR = DATA_DIR / "sample_images"
SRC_DIR = PROJECT_ROOT / "src"
API_DIR = PROJECT_ROOT / "api"
APP_DIR = PROJECT_ROOT / "app"
EVAL_DIR = PROJECT_ROOT / "eval"
OUTPUTS_DIR = PROJECT_ROOT / "outputs"
PROMPTS_DIR = PROJECT_ROOT / "prompts"
TESTS_DIR = PROJECT_ROOT / "tests"
OUTPUTS_DIR.mkdir(exist_ok=True)

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print("PROJECT_ROOT =", PROJECT_ROOT)
```

```python
import importlib.util
import tempfile
from datetime import datetime

paths = {
    "README.md": PROJECT_ROOT / "README.md",
    "src/inference.py": SRC_DIR / "inference.py",
    "src/guardrails.py": SRC_DIR / "guardrails.py",
    "src/preprocessing.py": SRC_DIR / "preprocessing.py",
    "src/database.py": SRC_DIR / "database.py",
    "src/metrics.py": SRC_DIR / "metrics.py",
    "api/main.py": API_DIR / "main.py",
    "app/streamlit_app.py": APP_DIR / "streamlit_app.py",
    "app/gradio_app.py": APP_DIR / "gradio_app.py",
    "eval/run_evaluation.py": EVAL_DIR / "run_evaluation.py",
    "tests/test_repository_smoke.py": TESTS_DIR / "test_repository_smoke.py",
    "data/synthetic_cases.csv": DATA_DIR / "synthetic_cases.csv",
    "data/sample_images": SAMPLE_IMAGES_DIR,
    "prompts": PROMPTS_DIR,
    "sql/schema.sql": PROJECT_ROOT / "sql" / "schema.sql",
}
file_checks = [{"path": k, "exists": v.exists()} for k, v in paths.items()]
deps = []
for package in ["pandas", "PIL", "matplotlib", "fastapi", "streamlit", "pytest"]:
    deps.append({"package": package, "available": importlib.util.find_spec(package) is not None})
DB_PATH = Path(tempfile.gettempdir()) / "assistant_radio_notebooks.sqlite"
file_checks, deps, DB_PATH
```

```python
report = OUTPUTS_DIR / "setup_check.txt"
lines = [f"generated_at={datetime.now().isoformat(timespec='seconds')}", f"PROJECT_ROOT={PROJECT_ROOT}", f"DB_PATH={DB_PATH}", "", "Files:"]
lines += [f"- {row['path']}: {'OK' if row['exists'] else 'MISSING'}" for row in file_checks]
lines += ["", "Dependencies:"]
lines += [f"- {row['package']}: {'OK' if row['available'] else 'MISSING'}" for row in deps]
report.write_text("\n".join(lines), encoding="utf-8")
print(report)
```
